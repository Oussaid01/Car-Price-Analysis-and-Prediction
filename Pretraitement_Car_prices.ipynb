# Introduction 
Dans un marché automobile en constante évolution, la compréhension des préférences des consommateurs et des dynamiques concurrentielles entre les marques revêt une importance cruciale pour les fabricants et les vendeurs. Le jeu de données "car_prices" constitue une mine d'informations précieuses permettant d'explorer en profondeur ces aspects fondamentaux du marché automobile contemporain.

# Objectif

#### Objectif principal de cette analyse :
Est de comparer la concurrence entre les différentes marques de voitures et de comprendre les préférences des consommateurs en matière de véhicules et de marques. En examinant de près les données fournies par le jeu de données, nous cherchons à dégager des insights significatifs sur les choix des consommateurs, ainsi que sur les tendances et les dynamiques concurrentielles qui régissent le marché automobile.

#### Objectif Commercial : 
On va faire une prediction du "SellingPrice" en fonction de quelques paramètres.

# Informations sur le dataset 'car_prices'

- **Year (année)** : L'année de fabrication du véhicule.  

- **Make (marque)** : La marque du véhicule, par exemple Kia, Toyota, etc.  

- **Model (modèle)** : Le modèle spécifique du véhicule, par exemple Sorento, Camry, etc.  

- **Trim (finition)** : La version spécifique ou la finition du modèle, par exemple LX, LE, etc.  

- **Body (type de carrosserie)** : Le type de carrosserie du véhicule, par exemple SUV, berline, etc.  

- **Transmission (transmission)** : Le type de transmission du véhicule, par exemple automatique ou manuelle.  

- **Vin (numéro d'identification du véhicule)** : Le numéro d'identification unique attribué au véhicule.  

- **State (état)** : L'état où la transaction du véhicule a eu lieu.  

- **Condition (état du véhicule)** : L'état général du véhicule, évalué sur une échelle de notation (par exemple de 1 à 5).  

- **Odometer (kilométrage)** : Le nombre de kilomètres parcourus par le véhicule.  

- **Color (couleur)** : La couleur extérieure du véhicule.  

- **Interior (intérieur)** : La couleur ou les caractéristiques de l'intérieur du véhicule.  

- **Seller (vendeur)** : Le vendeur du véhicule, par exemple un concessionnaire ou un particulier.  

- **Mmr (prix de gros recommandé)** : Le prix de gros recommandé pour le véhicule.  

- **Sellingprice (prix de vente)** : Le prix auquel le véhicule a été vendu.  

- **Saledate (date de vente)** : La date à laquelle la transaction de vente du véhicule a eu lieu.

# Prétraitement des données

```python
# Imporation des bibliothèques
import pandas as pd
import numpy as np
```

```python
# Lecture du fichier csv
csvFile = pd.read_csv("car_prices.csv",header=0)
#Mettre le fichier dans un DataFrame
df = pd.DataFrame(csvFile)
# Afficher les premièrs lignes
df.head()
```

```python
# afficher les derniers lignes
df.tail()
```

```python
# Afficher la dimension
df.shape
```

```python
# Afficher la liste des colonnes
print(df.columns)
```

```python
# Afficher le nombre de voiture dans chaque marque
df["make"].value_counts(dropna=True)
```

```python
# Remplacer 'tk' par 'truck' dans toute la colonne make
df['make'] = df['make'].str.replace('tk', 'truck')
```

```python
# Standardisation des noms de marques (pour éviter les doublons comme 'FORD' vs 'Ford')
df['make'] = df['make'].str.title()
print(df["make"])
```

```python
# Vérification de nombre de voiture dans chaque marque
df["make"].value_counts(dropna=True)
```

```python
# Afficher le type de chaque colonne
print(df.dtypes)
```

```python
# Informations sur les données
print(df.info())
```

```python
# Comptage des valeurs nulles par colonne
print(df.isnull().sum())
```

```python
# Remplacer les valeurs manquantes de model par le mode
df['model'] = df['model'].fillna(df['model'].mode()[0])
```

```python
# Remplacer les valeurs manquantes de trim par le mode
df['trim'] = df['trim'].fillna(df['trim'].mode()[0])
```

```python
# Remplacer les valeurs manquantes de body par le mode
df['body'] = df['body'].fillna(df['body'].mode()[0])
```

```python
# Remplacer les valeurs manquantes de transmission par le mode
df['transmission'] = df['transmission'].fillna(df['transmission'].mode()[0])
```

```python
# Remplacer les valeurs manquantes de condition par la médiane
df['condition'] = df['condition'].fillna(df['condition'].median())
```

```python
# Remplacer les valeurs manquantes de odometer par la médiane
df['odometer'] = df['odometer'].fillna(df['odometer'].median())
```

```python
# Remplacer les valeurs manquantes de color par le mode
df['color'] = df['color'].fillna(df['color'].mode()[0])
```

```python
# Remplacer les valeurs manquantes de interior par le mode
df['interior'] = df['interior'].fillna(df['interior'].mode()[0])
```

```python
# supprimer les lignes avec valeurs manquantes importantes
df = df.dropna(subset=['sellingprice', 'mmr','make'])
```

```python
# Afficher le nombre de valeurs uniques et les 10 types de carrosserie les plus fréquents
print(f"Nombre de types de carrosserie uniques : {df['body'].nunique()}")
display(df['body'].value_counts().head(10))
```

### Interprétation de la Variable `body`
**`body`** : Le dataset contient `86` types de carrosserie uniques. On observe une dominance de certains types, comme les berlines et les SUV, ce qui est cohérent avec le marché automobile. Les types moins fréquents pourraient représenter des niches ou des véhicules spécialisés. Il est également à noter la présence de types en majuscules et minuscules (ex: 'Sedan' et 'sedan'), ce qui pourrait nécessiter une standardisation ultérieure.

## Standardisation de la Colonne 'body'

```python
# Convertir la colonne 'body' en minuscules
df['body'] = df['body'].str.lower()

# Afficher le nombre de valeurs uniques et les 10 types de carrosserie les plus fréquents après standardisation
print("\nAnalyse de la colonne 'body' après standardisation :")
print(f"Nombre de types de carrosserie uniques : {df['body'].nunique()}")
display(df['body'].value_counts().head(10))
```

### Interprétation après Standardisation de 'body'

Après avoir converti toutes les entrées de la colonne `body` en minuscules, le nombre de types de carrosserie uniques a été réduit à **`45`**. Cela confirme que la standardisation a bien regroupé les catégories similaires, permettant une vue plus cohérente et simplifiée des types de carrosserie. Les décomptes des types les plus fréquents sont maintenant consolidés, ce qui facilitera les analyses futures.

```python
# Afficher le nombre de valeurs uniques et les 10 couleurs extérieures les plus fréquentes
print("\nAnalyse de la colonne 'color' :")
print(f"Nombre de couleurs uniques : {df['color'].nunique()}")
display(df['color'].value_counts().head(10))
```

```python
# Afficher le nombre de valeurs uniques et les 10 couleurs intérieures les plus fréquentes
print("\nAnalyse de la colonne 'interior' :")
print(f"Nombre de couleurs intérieures uniques : {df['interior'].nunique()}")
display(df['interior'].value_counts().head(10))
```

### Interprétation des Variables Catégorielles (`color`, `interior`)

Il faudra vérifier s'il y a des incohérences de casse ou des synonymes (ex: 'black' et 'BLK') qui nécessiteraient une standardisation, comme pour la colonne `body`.

```python
# Remplacer '—' par 'unknown' dans la colonne 'interior'
df['interior'] = df['interior'].replace('—', 'unknown')

# Afficher le nombre de valeurs uniques et les 10 couleurs intérieures les plus fréquentes après standardisation
print("\nAnalyse de la colonne 'interior' après standardisation :")
print(f"Nombre de couleurs intérieures uniques : {df['interior'].nunique()}")
display(df['interior'].value_counts().head(10))
```
