## Analyse de concurrence entre les marques

```python
# Sélectionner les 10 marques les plus représentées
top_10_makes = df['make'].value_counts().nlargest(10).index

# Filtrer le dataframe pour inclure uniquement les 10 marques les plus représentées
df_top_10 = df[df['make'].isin(top_10_makes)]

# Visualiser la répartition des 10 marques de voitures les plus représentées
plt.figure(figsize=(10, 6))
sns.countplot(y='make', data=df_top_10, order=df_top_10['make'].value_counts().index, palette="viridis")
plt.title('Répartition des 10 Marques de Voitures les Plus Représentées')
plt.xlabel('Nombre de Voitures')
plt.ylabel('Marque')
plt.show()
```

### Interprétation : Répartition des 10 Marques de Voitures les Plus Représentées
Le graphique à barres horizontal ci-dessus présente la répartition des 10 marques de voitures les plus représentées dans notre jeu de données. Ces marques sont celles qui ont le plus grand nombre de voitures enregistrées. En observant cette visualisation, nous pouvons identifier les marques dominantes sur le marché automobile, ainsi que leur relative popularité parmi les consommateurs. Il est intéressant de noter que [Chevrolet] est en tête en termes de nombre de voitures, suivi de près par [Ford]. Cette information nous donne un aperçu initial des marques les plus importantes à considérer dans notre analyse de la compétition entre les marques et des préférences des consommateurs.

### Prix de Vente Moyens par Marque (Top 12)

```python
# Calculer le prix de vente moyen par marque
mean_selling_price_by_make = df.groupby('make')['sellingprice'].mean().reset_index()

# Trier les marques par prix de vente moyen décroissant et sélectionner le top 12
top_12_makes_by_price = mean_selling_price_by_make.sort_values(by='sellingprice', ascending=False).head(12)
display(top_12_makes_by_price.head(12))

# Créer un graphique à barres horizontal avec seaborn
plt.figure(figsize=(12, 7))
sns.barplot(x='sellingprice', y='make', data=top_12_makes_by_price, palette='viridis')
plt.title('Prix de Vente Moyens des 12 Marques de Voiture les Plus Chères')
plt.xlabel('Prix de Vente Moyen ($)')
plt.ylabel('Marque')
plt.grid(axis='x')
plt.tight_layout()
plt.show()
```

## Interprétation des Prix de Vente Moyens par Marque

Le graphique à barres horizontal nous montre les 12 marques de voitures avec les prix de vente moyens les plus élevés dans notre dataset. Cette visualisation met en évidence plusieurs points importants.

**Les marques de luxe/premium :** les marques en tête sont des constructeurs de luxe reconnus tels que Rolls-Royce (~153 488 $), Ferrari (~127 210 $), Lamborghini (~112 625 $), Bentley (~74 367 $), et Tesla (~67 054 $). Cela est tout à fait cohérent avec leur positionnement sur le marché et leur valeur de revente élevée.

**La disparité des prix :** il y a une différence significative entre le prix moyen des marques les plus chères et celles juste en dessous. Par exemple, Rolls-Royce a un prix moyen près de cinq fois supérieur à celui de Land Rover (~31 617 $), qui clôture ce top 12.

**La présence de marques spécialisées :** on observe des marques comme Airstream (un constructeur de caravanes, potentiellement inclus via des véhicules motorisés ou d'occasion très haut de gamme) et Fisker, qui apparaissent dans ce top, suggérant une niche de marché ou des véhicules spécifiques à forte valeur.

Ces informations sont précieuses pour comprendre le segment de marché couvert par notre dataset et comment la marque influence directement la valeur de revente du véhicule. Pour les modèles prédictifs, la marque sera une caractéristique très influente.

### Répartition des Ventes par Marque (Dernières 3 années)

```python
# Convert 'saledate' to datetime objects
if not pd.api.types.is_datetime64_any_dtype(df['saledate']):
    df['saledate'] = pd.to_datetime(df['saledate'], errors='coerce', utc=True)

# Filtrer les données pour les 3 dernières années
# Déterminer l'année de vente maximale dans le dataset
max_sale_year = df['saledate'].dt.year.max()

# Filtrer les données pour les 3 dernières années (max_sale_year, max_sale_year-1, max_sale_year-2)
last_3_years_data = df[df['saledate'].dt.year >= (max_sale_year - 2)]

# Compter le nombre de voitures vendues par marque pour les 3 dernières années
cars_sold_by_make_last_3_years = last_3_years_data['make'].value_counts()

# Sélectionner les trois premières marques les plus dominantes
top_3_makes_last_3_years = cars_sold_by_make_last_3_years.head(3)
```

```python
# Afficher les trois premières marques les plus dominantes
print("Top 3 marques les plus dominantes sur les 3 dernières années :\n")
display(top_3_makes_last_3_years)
```

```python
# Préparer les données pour le diagramme circulaire
makes_pie = top_3_makes_last_3_years.index.tolist()
cars_sold_pie = top_3_makes_last_3_years.values.tolist()

# Créer un diagramme circulaire
plt.figure(figsize=(8, 8))
plt.pie(cars_sold_pie, labels=makes_pie, autopct='%1.1f%%', startangle=140, colors=['#66b3ff', '#99ff99', '#ffcc99'])
plt.title('Répartition des Ventes par Marque (Dernières 3 années)')
plt.show()
```

```python
plt.figure(figsize=(10, 6)) # Créer une nouvelle figure pour le bar plot
sns.barplot(x=top_3_makes_last_3_years.index, y=top_3_makes_last_3_years.values, palette="viridis", hue=top_3_makes_last_3_years.index, legend=False)
plt.title('Nombre de Voitures Vendues par Marque (Dernières 3 années)')
plt.xlabel('Marque')
plt.ylabel('Nombre de Voitures Vendues')
plt.xticks(rotation=45)
plt.grid(axis='y')

plt.tight_layout()
plt.show()
```

### Interprétation de la Répartition des Ventes par Marque (Dernières 3 années)

Ce diagramme circulaire visualise la part de marché des trois marques de voitures les plus vendues au cours des trois dernières années couvertes par le dataset. Il nous permet de :

*   **Identifier les leaders du marché récent** : Il met en évidence les marques qui dominent les ventes les plus récentes, ce qui peut indiquer des tendances actuelles du marché ou des préférences des consommateurs.
*   **Quantifier la dominance** : L'autopct sur le diagramme circulaire montre directement le pourcentage de ventes que chaque marque représente parmi les trois premières, offrant une vue claire de la concentration du marché.

# <font color=red>  Nuage de points et Coefficient de corrélation de Pearson

```python
from scipy import stats
```

### Relation entre le Kilométrage (`odometer`) et le Prix de Vente (`sellingprice`)

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='odometer', y='sellingprice', data=df, alpha=0.6, color='darkgreen')
plt.title('Relation entre le Kilométrage et le Prix de Vente')
plt.xlabel('Kilométrage (Odometer)')
plt.ylabel('Prix de Vente')
plt.grid(True)
plt.show()
```

### Interprétation de la Relation Kilométrage vs. Prix de Vente

Le nuage de points devrait montrer une **corrélation négative** : plus le kilométrage est élevé, plus le prix de vente a tendance à être bas. On s'attend à voir une densité plus forte de points aux kilométrages plus faibles et des prix plus élevés, et une dispersion des prix à la baisse à mesure que le kilométrage augmente. Des exceptions ou des "outliers" (véhicules à faible kilométrage mais prix bas, ou vice-versa) pourraient également apparaître.

```python
# Coefficient de Pearson entre les variables 'odometer' 'sellingprice'

stats.pearsonr(df['odometer'], df['sellingprice'])
```

### Interprétation du Coefficient de Pearson entre 'odometer' et 'sellingprice'

Le coefficient de corrélation de Pearson entre le kilométrage (`odometer`) et le prix de vente (`sellingprice`) est d'environ **-0.58** pour une p-valeur égale à 0. Voici ce que cela signifie :

*   **Type de Corrélation** : La valeur négative indique une **corrélation négative** entre ces deux variables. Cela signifie que, généralement, à mesure que le kilométrage d'un véhicule augmente, son prix de vente a tendance à diminuer.

*   **Force de la Corrélation** : Une valeur de -0.58 représente une **corrélation modérée à forte**. Ce n'est pas une corrélation parfaite (-1), mais elle est suffisamment significative pour indiquer que le kilométrage est un facteur important influençant le prix de vente d'une voiture. La relation est claire : plus une voiture a roulé, moins elle est chère.

Cette observation est cohérente avec les attentes du marché automobile, où le kilométrage est un indicateur clé de l'usure et de la valeur résiduelle d'un véhicule.

### Relation entre le MMR (`mmr`) et le Prix de Vente (`sellingprice`)

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='mmr', y='sellingprice', data=df, alpha=0.6, color='darkviolet')
plt.title('Relation entre le MMR et le Prix de Vente')
plt.xlabel('MMR (Manheim Market Report)')
plt.ylabel('Prix de Vente')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='mmr', y='sellingprice', data=df, scatter_kws={'alpha':0.3}, line_kws={'color':'red'})
plt.title('Régression Linéaire : MMR vs Prix de Vente')
plt.xlabel('MMR (Manheim Market Report)')
plt.ylabel('Prix de Vente')
plt.grid(True)
plt.show()
```

### Interprétation de la Relation MMR vs. Prix de Vente

Comme prévu, le nuage de points montre une **corrélation quasi-parfaite et positive** entre le MMR et le prix de vente. Les points s'alignent très étroitement le long d'une ligne, confirmant que le MMR est un excellent indicateur du prix auquel les véhicules sont vendus. Il est normal qu'il y ait une légère dispersion, car le MMR est une valeur estimée et le prix de vente réel peut varier en fonction d'autres facteurs et des négociations finales.
