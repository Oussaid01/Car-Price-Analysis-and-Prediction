### Répartition des Ventes par Marque (Dernières 3 années)

```python
# Afficher les trois premières marques les plus dominantes
print("Top 3 marques les plus dominantes sur les 3 dernières années :\n")
display(top_3_makes_last_3_years)
```

### Relation entre le Kilométrage (`odometer`) et le Prix de Vente (`sellingprice`)

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='odometer', y='sellingprice', data=df, alpha=0.6, color='darkgreen')
plt.title('Relation entre le Kilométrage et le Prix de Vente')
plt.xlabel('Kilométrage (Odometer)')
plt.ylabel('Prix de Vente')
plt.grid(True)
plt.show()
```

### Interprétation de la Relation Kilométrage vs. Prix de Vente

Le nuage de points devrait montrer une **corrélation négative** : plus le kilométrage est élevé, plus le prix de vente a tendance à être bas. On s'attend à voir une densité plus forte de points aux kilométrages plus faibles et des prix plus élevés, et une dispersion des prix à la baisse à mesure que le kilométrage augmente. Des exceptions ou des "outliers" (véhicules à faible kilométrage mais prix bas, ou vice-versa) pourraient également apparaître.

```python
# Coefficient de Pearson entre les variables 'odometer' 'sellingprice'

stats.pearsonr(df['odometer'], df['sellingprice'])
```

### Interprétation du Coefficient de Pearson entre 'odometer' et 'sellingprice'

Le coefficient de corrélation de Pearson entre le kilométrage (`odometer`) et le prix de vente (`sellingprice`) est d'environ **-0.58** pour une p-valeur égale à 0. Voici ce que cela signifie :

*   **Type de Corrélation** : La valeur négative indique une **corrélation négative** entre ces deux variables. Cela signifie que, généralement, à mesure que le kilométrage d'un véhicule augmente, son prix de vente a tendance à diminuer.

*   **Force de la Corrélation** : Une valeur de -0.58 représente une **corrélation modérée à forte**. Ce n'est pas une corrélation parfaite (-1), mais elle est suffisamment significative pour indiquer que le kilométrage est un facteur important influençant le prix de vente d'une voiture. La relation est claire : plus une voiture a roulé, moins elle est chère.

Cette observation est cohérente avec les attentes du marché automobile, où le kilométrage est un indicateur clé de l'usure et de la valeur résiduelle d'un véhicule.

### Analyse de la Corrélation Linéaire entre l'Année, l'État, le Kilométrage, le MMR et le Prix de Vente

La carte de corrélation montre que la variable MMR possède une très forte corrélation positive avec le prix de vente (0.98), ce qui signifie que lorsque la valeur MMR augmente, le prix de vente augmente également.

On remarque aussi une corrélation négative entre le kilométrage et le prix de vente (-0.58). Cela indique que les véhicules ayant un kilométrage élevé ont généralement un prix plus faible.

L’année du véhicule présente une corrélation positive avec le prix de vente (0.59), ce qui signifie que les véhicules les plus récents sont souvent plus chers.

Enfin, la variable condition possède une faible corrélation positive avec le prix de vente (0.31), ce qui montre qu’un meilleur état du véhicule peut influencer légèrement le prix.

```python
# Carte des Corrélations Améliorée

plt.figure(figsize = (10, 6))

mask = np.triu(np.ones_like(
    df[['year', 'condition', 'odometer', 'mmr', 'sellingprice']].corr(),
    dtype=bool))

sns.heatmap(
    df[['year', 'condition', 'odometer', 'mmr', 'sellingprice']].corr(),
    mask=mask,
    center=0,
    cmap='RdBu',
    linewidths=1,
    annot=True,
    fmt=".2f",
    vmin=-1,
    vmax=1)

plt.title('Carte des Corrélations', fontsize=15, fontweight="bold")

plt.show()
```

La carte des corrélations améliorée permet de visualiser plus clairement les relations entre les variables quantitatives du dataset.

Le kilométrage présente une corrélation négative avec le prix de vente (-0.58), ce qui signifie que les véhicules ayant parcouru plus de kilomètres ont tendance à être moins chers.

L’année du véhicule possède également une corrélation positive avec le prix de vente (0.59), montrant que les voitures récentes sont généralement vendues à des prix plus élevés.

### Coefficient de Pearson entre 'odometer' et 'sellingprice' selon le type de transmission

```python
# Coefficient de Pearson entre 'odometer' et 'sellingprice' selon le type de transmission
print(
    'Automatic :',
    stats.pearsonr(
        df[df['transmission'] == 'automatic']['odometer'],
        df[df['transmission'] == 'automatic']['sellingprice']
    )
)

print(
    'Manual :',
    stats.pearsonr(
        df[df['transmission'] == 'manual']['odometer'],
        df[df['transmission'] == 'manual']['sellingprice']
    )
)
```

Les résultats du coefficient de Pearson montrent une corrélation négative entre le kilométrage et le prix de vente pour les véhicules automatiques et manuels.

Pour les véhicules automatiques, le coefficient est de -0.58, tandis que pour les véhicules manuels il est de -0.60. Cela signifie que lorsque le kilométrage augmente, le prix de vente diminue dans les deux types de transmission.

La corrélation est légèrement plus forte pour les véhicules manuels, ce qui indique que le kilométrage influence un peu plus le prix des voitures manuelles.

Enfin, les p-values sont égales à 0.0, ce qui confirme que les corrélations obtenues sont statistiquement significatives.

### Analyse de la Corrélation des Couleurs avec le Prix de Vente

```python
# Analyse de la Corrélation des Couleurs avec le Prix de Vente

# Suppression des couleurs inconnues
df_colors = df[df['color'] != 'unknown']

# Encodage des couleurs en variables numériques
color_encoded = pd.get_dummies(df_colors['color'])

# Calcul de la corrélation avec le prix de vente
color_corr = color_encoded.corrwith(df_colors['sellingprice'])

# Sélection des 5 couleurs les plus corrélées
top_5_corr = color_corr.abs().sort_values(ascending=False).head(5)

# Diagramme circulaire
plt.figure(figsize=(10, 6))

plt.pie(
    top_5_corr.values,
    labels=top_5_corr.index,
    autopct='%1.1f%%'
)

plt.title(
    'Corrélation des Cinq Couleurs les Plus Influentes sur le Prix de Vente'
)

plt.axis('equal')

plt.show()
```

### Interprétation

Ce diagramme circulaire présente les cinq couleurs ayant la plus forte corrélation avec le prix de vente des véhicules.

On remarque que la couleur black possède la corrélation la plus importante avec le prix de vente, suivie par les couleurs gold, green, silver et white.

Cependant, les différences entre les pourcentages restent relativement faibles, ce qui montre que la couleur influence légèrement le prix des véhicules.

Ces résultats indiquent que certaines couleurs peuvent être plus appréciées sur le marché automobile, mais que leur impact reste limité comparativement à d’autres variables comme le MMR, le kilométrage ou l’année du véhicule.
